```python
# reading all the files from instacart folder
orders = pd.read_csv(BASE_PATH + 'orders.csv')
aisles = pd.read_csv(BASE_PATH + 'aisles.csv')
departments = pd.read_csv(BASE_PATH + 'departments.csv')
order_products_prior = pd.read_csv(BASE_PATH + 'order_products__prior.csv')
order_products_train = pd.read_csv(BASE_PATH + 'order_products__train.csv')
products = pd.read_csv(BASE_PATH + 'products.csv')
```

```python
# Check for null values across all 6 tables before cleaning
print("orders:",orders.isnull().sum())
print("aisles:",aisles.isnull().sum())
print("departments:",departments.isnull().sum())
print("order_products_prior:",order_products_prior.isnull().sum())
print("order_products_train:",order_products_train.isnull().sum())
print("products:",products.isnull().sum())
```

```python
# Check for duplicate rows across all 6 tables
print("orders:",orders.duplicated().sum())
print("aisles:",aisles.duplicated().sum())
print("departments:",departments.duplicated().sum())
print("order_products_prior:",order_products_prior.duplicated().sum())
print("order_products_train:",order_products_train.duplicated().sum())
print("products:",products.duplicated().sum())
```

```python
# Checking the data inside the column to see whats relavant or if anything needs to be filtered out
orders['eval_set'].value_counts()
```

```python
# Filtering out unwanted data from the column eval_set
orders_filtered = orders[orders['eval_set'].isin(['prior', 'train'])]
```

```python
# Checking the new dataframe with filtered data
orders_filtered['eval_set'].value_counts()
```

```python
orders_filtered.head()
```

```python
order_products_prior.head()
```

```python
# Merge order-product-prior data with order details using order_id as the key
orders_products_merged = pd.merge(order_products_prior, orders_filtered, on='order_id')
print("orders_products_merged:", orders_products_merged.shape)
```

```python
# Merge order-product-train data with order details using order_id as the key
orders_products_train_merged = pd.merge(order_products_train, orders_filtered, on='order_id')
print("orders_products_train_merged:", orders_products_train_merged.shape)
```

```python
# Stack prior and train order-product tables into one master table
all_order_products = pd.concat([orders_products_merged, orders_products_train_merged])
print("all_order_products:", all_order_products.shape)
```

```python
aisles.head()
```

```python
departments.head()
```

```python
# Enrich products table with aisle and department names
products_enriched = pd.merge(products, aisles, on='aisle_id')
products_enriched = pd.merge(products_enriched, departments, on='department_id')
print("products_enriched:", products_enriched.shape)
```

```python
products_enriched.head()
```

```python
all_order_products.head()
```

```python
# Final master table combining all order-product data with product, aisle, and department details
master = pd.merge(all_order_products, products_enriched, on='product_id')
print("master:", master.shape)
```

```python
master.head()
```

```python
# Aggregate order history to one row per customer
customer_summary = (
    orders_filtered
    .groupby('user_id')
    .agg(
        total_orders     = ('order_number', 'max'),
        avg_days_between = ('days_since_prior_order', 'mean'),
        std_days_between = ('days_since_prior_order', 'std'),
        max_gap          = ('days_since_prior_order', 'max')
    )
    .reset_index()
)
print("customer_summary:", customer_summary.shape)
```

```python
customer_summary.head()
```

```python
# Fill NaN std with 0 — customers with only 1 order have no variability
customer_summary['std_days_between'] = customer_summary['std_days_between'].fillna(0)
```

```python
# Checking for nulls
print(customer_summary.isnull().sum())
```

```python
import os
os.makedirs(BASE_PATH + 'cleaned', exist_ok=True)
```

```python
# Save cleaned files to Drive so we don't have to reload raw data every session
customer_summary.to_csv(BASE_PATH + 'cleaned/customer_summary.csv', index=False)
orders_filtered.to_csv(BASE_PATH + 'cleaned/orders_clean.csv', index=False)
products_enriched.to_csv(BASE_PATH + 'cleaned/products_enriched.csv', index=False)
```
